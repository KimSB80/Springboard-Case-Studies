#### i. Deal with missing values

```python
# Examine missing values
data[data['phone'].isnull()]
```

```python
# Examine missing values
data[data['avg_rating_by_driver'].isnull()]
```

Since the total percent of missing values in the data is only 1.3%, we can probably just drop these values. The most concerning of these is the missing values in the 'avg_rating_of_driver' column, representing ~16% of that column's data. However, we don't have a good way to interpolate or infer these missing values since it's a rating by the customer, and any assumptions on interpolating this value could skew the results. Therefore, I think the best decision is to drop these rows since we still have a good amount of data even without them.<br>
A caveat to this decision is that there may be some bias in people who don't leave a review -- perhaps if they had a very average experience they might not be inspired to leave any review. This is something to consider when analyzing the overall usefulness of the model. 

```python
# Drop rows with missing values
df = data.dropna()
#df.isna().sum()
```

#### ii. Create predictor column

```python
# Find date the data was pulled
today = pd.to_datetime(max(df['last_trip_date']))
print('The data was pulled on', today)

# Find users who were active in the past 30 days
days_elapsed =  (today - df['last_trip_date']).dt.days

# Create predictor column (active)
df.loc[days_elapsed <= 30, 'active'] = 1
df.loc[days_elapsed > 30, 'active'] = 0
df.head()
```

```python
# Calculate the fraction of the observed users that were retained 
retained = df['active'].mean()
print('The total percent of users retained was {:.2f}%'.format(round(retained*100, 2)))
```

There is a slight imbalance in class data for the predictor variable, but not bad. If we decide this needs to be dealt with later, we can use resampling -- oversamping (of the minority class) or undersampling (of the majority class).

```python
# Get columns
df_log_columns = num_columns.drop('weekday_pct')

# Create new dataframe
df_log = pd.DataFrame()
offset = 1  # Small constant to avoid taking log(0)

for c in df_log_columns:
    new_col_name = f"log_{c}"
    df_log[new_col_name] = np.log10(df[c].values + offset)
#    df.loc[:, new_col_name] = np.log10(df[c])

df_log.boxplot(figsize=(15,15))
plt.subplots_adjust(hspace=0.5);
```

For now, I'll choose the log transform with shifted data since this helps with the skew. There are many choices for transformations -- the main concern here is dealing with the skew and negative values created by a log transformation of 0.

## 2. Modeling Part 1

```python
# Load packages
from scipy import stats

from sklearn.preprocessing import StandardScaler, MinMaxScaler, scale, OneHotEncoder, FunctionTransformer
from sklearn.model_selection import train_test_split, cross_validate, GridSearchCV
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, f1_score, recall_score, classification_report, confusion_matrix, ConfusionMatrixDisplay

from sklearn.compose import make_column_selector as selector
from sklearn.compose import ColumnTransformer
```

```python
# Load saved data
dfmod = pd.read_csv('./clean_data.csv', index_col=[0])
dfmod.head()
```

### Model Training and Selection
This is a binary classification problem, so we'll try out four different classification models first to see how they do. 
1. Logistic regression
2. Random forest
3. Gradient boosting
4. SVM

```python
dfmod.shape
```

```python
# Define target (y-variable)
target_name = "active"
target = dfmod[target_name]

dat = dfmod.drop(columns=[target_name])
dat.info()
```

```python
# Select columns based on data type (categorical)
numerical_columns_selector = selector(dtype_exclude=[object,bool])
categorical_columns_selector = selector(dtype_include=[object, bool])

numerical_features = numerical_columns_selector(dat) # take log of these features
categorical_features = categorical_columns_selector(dat) # dummy encode these features
print('Numerical features:', numerical_features)
print('Categorical features:', categorical_features)
```

```python
# Create the preprocessors for the numerical and categorical columns 
preprocessor = ColumnTransformer(
    transformers=[
        ('log', FunctionTransformer(np.log1p, validate=False), numerical_features), 
        ('one-hot-encoder', OneHotEncoder(handle_unknown="ignore"), categorical_features)],
    remainder='passthrough')
```

#### Train-test split data

```python
# First split the data into train and test set
X_train, X_test, y_train, y_test = train_test_split(dat, target,
                                            test_size=0.3, 
                                            random_state=43)
```

```python
# Check shape of data
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

#### 1. Logistic Regression Model

```python
# Instantiate model
classifier = LogisticRegression()
#classifier = LogisticRegression(solver = 'liblinear', max_iter = 500, C = 1000)

# Fit model to training data
lrPipe = make_pipeline(preprocessor,
    LogisticRegression()
)
```

```python
# Fit model
lr_mod = lrPipe.fit(X_train, y_train)

# Get predicions
y_predict_training = lr_mod.predict(X_train)
y_predict_test = lr_mod.predict(X_test)
```

```python
print("[Training] Accuracy score:", 
      f'{accuracy_score(y_train, y_predict_training):.2f}')
```

```python
# Classification report for training data
print(classification_report(y_train, y_predict_training))
```

```python
# Assess basic regression model: accuracy and recall for training and test data
pd.DataFrame([
    [lr_mod.score(X_train, y_train), recall_score(y_train, lr_mod.predict(X_train))],
    [lr_mod.score(X_test, y_test), recall_score(y_test, lr_mod.predict(X_test))],
], columns = ['Accuracy', 'Recall'], index=['Train', 'Test'])
```

#### 2. Random forest model

```python
# Create a pipeline 
rfPipe = make_pipeline(preprocessor,
    RandomForestClassifier(
    n_estimators=100,           # Number of trees in the forest
    max_depth=5,              # Maximum depth of individual trees (None means unlimited depth)
    min_samples_leaf=3,          # Minimum number of samples required to be at a leaf node
    max_features='auto',         # Number of features to consider when looking for the best split ('auto' uses all features)
    bootstrap=True,              # Whether bootstrap samples are used when building trees
    random_state=43,             # Random seed for reproducibility
    n_jobs=-1                    # Number of CPU cores to use for parallel processing (-1 means use all available cores)
)
)
```

```python
# Fit model
rf_mod = rfPipe.fit(X_train, y_train)

# Get predicions
y_predict_training = rf_mod.predict(X_train)
y_predict_test = rf_mod.predict(X_test)
```

```python
# Assess model
print("[Training] Accuracy score:", 
      f'{accuracy_score(y_train, y_predict_training):.2f}')
```

```python
# Classification report for training data
print(classification_report(y_train, y_predict_training))
```

```python
# Assess basic forest model: accuracy and recall for training and test data
pd.DataFrame([
    [rf_mod.score(X_train, y_train), recall_score(y_train, rf_mod.predict(X_train))],
    [rf_mod.score(X_test, y_test), recall_score(y_test, rf_mod.predict(X_test))],
], columns = ['Accuracy', 'Recall'], index=['Train', 'Test'])
```

This base model did very well on the training data (to the point of overfitting), and okay on the test data. I modified the hyperparameters a bit to deal with overfitting.

#### 3. Gradient boosting model

```python
# Create a pipeline 
gbPipe = make_pipeline(preprocessor,
    GradientBoostingClassifier()
)
```

```python
# Fit model
gb_mod = gbPipe.fit(X_train, y_train)

# Get predicions
y_predict_training = gb_mod.predict(X_train)
y_predict_test = gb_mod.predict(X_test)
```

```python
# Assess model
print("[Training] Accuracy score:", 
      f'{accuracy_score(y_train, y_predict_training)}')
```

```python
# Classification report for training data
print(classification_report(y_train, y_predict_training))
```

```python
# Assess basic boosting model: accuracy and recall for training and test data
pd.DataFrame([
    [gb_mod.score(X_train, y_train), recall_score(y_train, y_predict_training)],
    [gb_mod.score(X_test, y_test), recall_score(y_test, y_predict_test)],
], columns = ['Accuracy', 'Recall'], index=['Train', 'Test'])
```

#### 4. SVM model

```python
# Create a pipeline 
svmPipe = make_pipeline(preprocessor,
    SVC()
)
```

```python
# Fit model
svm_mod = svmPipe.fit(X_train, y_train)

# Get predicions
y_predict_training = svm_mod.predict(X_train)
y_predict_test = svm_mod.predict(X_test)
```

```python
# Assess model
print("[Training] Accuracy score:", 
      f'{accuracy_score(y_train, y_predict_training)}')
```

```python
# Classification report for training data
print(classification_report(y_train, y_predict_training))
```

```python
# Assess basic svm model: accuracy and recall for training and test data
pd.DataFrame([
    [gb_mod.score(X_train, y_train), recall_score(y_train, y_predict_training)],
    [gb_mod.score(X_test, y_test), recall_score(y_test, y_predict_test)],
], columns = ['Accuracy', 'Recall'], index=['Train', 'Test'])
```

##### Model summary:
Of the four models tested so far, the gradient boosting model performed best in terms of the metrics assessed. The random forest model did well as well, but appeared to overfit the training data and therefore not do as well on the test data. It could be worth working with that model more, but the gradient boosting model performed best here so we'll move ahead with that first.


## Modeling - Part 2
Use hyperparameter tuning and cross-validation to fit the gradient boosting model to the data. Then assess how it performs on the test data.

```python
# Create a pipeline 
gbPipe = make_pipeline(preprocessor,
    GradientBoostingClassifier()
)
```

```python
gbPipe.get_params().keys()
```

```python
# Filter out warnings
warnings.filterwarnings("ignore")

param_grid = {
        'gradientboostingclassifier__n_estimators' : [500,1000],
        'gradientboostingclassifier__max_depth': [3, 5],
        'gradientboostingclassifier__max_features': [1.0, 'sqrt'],
        'gradientboostingclassifier__learning_rate': [0.1, 0.5],
        'gradientboostingclassifier__random_state': [43]
        }

gb_grid_cv = GridSearchCV(gbPipe, param_grid = param_grid, cv=5, n_jobs= -1) # n_jobs=-1 means running jobs in parallel
gb_grid_cv.fit(X_train,y_train)
print(gb_grid_cv.best_estimator_)
```

```python
print("Best Parameters: \n{}\n".format(gb_grid_cv.best_params_))
print("Best Test Score: \n{}\n".format(gb_grid_cv.best_score_))
```

#### Build final model
Use best params from the output of the hyperparameter tuning using 5-fold cross validation.

```python
# Process the data using ColumnTransformer
X_tr = preprocessor.fit_transform(X_train)
X_te = preprocessor.transform(X_test)
y_tr = y_train
y_te = y_test
```

```python
# Fit model with selected hyperparameters
gb_final = GradientBoostingClassifier(max_depth=3, n_estimators=1000, max_features='sqrt', random_state=43)

# Fit and predict
gb_final.fit(X_tr, y_tr)
y_pred = gb_final.predict(X_te)
```

```python
print("Accuracy score (train): {0:.3f}".format(gb_final.score(X_tr, y_tr)))
print("Accuracy score (test): {0:.3f}".format(gb_final.score(X_te, y_te)))
```

```python
# Classification report
target_names = ['Not Retained', 'Retained']
print(classification_report(y_te, y_pred, target_names=target_names))
```

```python
confusion_matrix(y_te, y_pred)
```

```python
cm = confusion_matrix(y_te, y_pred)
display_cm = ConfusionMatrixDisplay(confusion_matrix = cm, 
                                    display_labels = target_names)

fig, ax = plt.subplots(figsize=(4,4))
display_cm.plot(ax=ax)
plt.show()
```

```python
from sklearn.metrics import roc_auc_score, roc_curve, auc

# ROC curve
fpr, tpr, thresholds =  roc_curve(y_te, y_pred)
plt.plot(fpr,tpr)
plt.xlabel('1-Specificity(False Positive Rate)')
plt.ylabel('Sensitivity(True Positive Rate)')
plt.title('Receiver Operating Characteristic')
plt.plot([0, 1], [0, 1],'r--');

auc_score = auc(fpr, tpr)
print('AUC: %.3f' % auc_score)
```

## Visualize feature importance

```python
# Extract feature names from ColumnTransformer
feature_names = []
for name, transformer, features in preprocessor.transformers_:
    if hasattr(transformer, 'get_feature_names_out'):
        feature_names.extend(
            [f"{name}_{f}" for f in transformer.get_feature_names_out(features)])
    else:
        feature_names.extend(features)

# Replace substring in list of strings
f1 = [x.replace("standard_scaler_", "") for x in feature_names]
f2 = [x.replace("one-hot-encoder_", "") for x in f1]

print(f2)
```

```python
features = f2
importances = gb_final.feature_importances_
indices = np.argsort(importances)

fig = plt.figure(figsize=(12, 6))
plt.title('Feature Importances')
plt.barh(range(len(indices)), importances[indices], color='b', align='center')
plt.yticks(range(len(indices)), [features[i] for i in indices])
plt.xlabel('Relative Importance')
plt.show()
```
