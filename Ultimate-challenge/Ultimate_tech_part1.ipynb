## PART 1: EDA


### Import Packages and Load Data

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
logins = pd.read_json("logins.json")
```

```python
logins.head()
```

```python
logins.info()
```

There are no missing values in the data.

### Create new dataframe

```python
# create new timestamp in 15 min intervals
logins['login_floor'] = logins['login_time'].dt.floor('15T')
logins.head()
#len(logins['login_floor'].unique()) # checking for number of unique 15min interval bins
```

```python
# create new dataframe with a count number of logins per 15 min interval 
logins_15 = pd.DataFrame(logins.groupby(by='login_floor',as_index=False, sort=False).count())
logins_15.rename(columns={'login_floor':'timestamps','login_time':'login_counts'}, inplace=True)
logins_15.head()
```

```python
logins_15.describe(datetime_is_numeric=True)
```

That data quality looks good -- there are no negative values for number of logins, and the min and max are reasonable numbers.

#### Plot the data

```python
# create time series plot

# setting the dimensions of the plot
fig, ax = plt.subplots(figsize=(8, 8))
 
sns.lineplot(x = "timestamps", y = "login_counts", data = logins_15)
plt.ylabel("Number of logins")
plt.xticks(rotation = 25);
```

The timeseries above shows login counts every 15 minutes. Given the density of the data, it's hard to discern any patterns. I'll aggregate the data at larger time intervals to look for patterns.

### i. Aggregate data at larger time intervals
Aggregate at hourly, daily, and monthly intervals, and also look for patterns based on business hours or weekdays.

```python
data=logins_15

# Aggregation levels (hourly, daily, business hours, weekends)
data['hour'] = data['timestamps'].dt.hour
data['day'] = data['timestamps'].dt.date
data['month'] = data['timestamps'].dt.month
data['is_business_hour'] = (data['timestamps'].dt.hour >= 9) & (data['timestamps'].dt.hour < 17)
data['is_weekday'] = data['timestamps'].dt.weekday < 5
data.head()
```

```python
# Plot MEAN hourly, daily, and monthly aggretations
fig, axs = plt.subplots(3, 1, figsize=(8, 10))

sns.lineplot(x = "hour", y = "login_counts", data = data, ax=axs[0])
sns.lineplot(x = "day", y = "login_counts", data = data, ax=axs[1])
sns.lineplot(x = "month", y = "login_counts", data = data, ax=axs[2])


plt.ylabel("Number of logins")
axs[1].tick_params(axis='x', rotation=45)

# Set plot titles and labels
axs[0].set_title('Mean Hourly Logins')
axs[1].set_title('Mean logins by day')
axs[2].set_title('Mean logins by month')

for ax in axs:
    ax.set_xlabel('Time')
    ax.set_ylabel('Login Counts')
#    ax.legend()

plt.tight_layout()
plt.show()
```

```python
# Plot TOTAL hourly and daily aggregation
fig, axs = plt.subplots(2, 1, figsize=(8, 8))

daily_agg = data.groupby('day')['login_counts'].sum()
daily_agg.plot(ax=axs[0], color='green', label='Daily Summed Aggregation', rot=45)

# Plot monthly aggregation
daily_agg = data.groupby('month')['login_counts'].sum()
daily_agg.plot(ax=axs[1], color='green', label='Monthly Summed Aggregation', rot=45)

# Set plot titles and labels
axs[0].set_title('Total Logins by Hour')
axs[1].set_title('Total Logins by Day')

for ax in axs:
    ax.set_xlabel('Time')
    ax.set_ylabel('Login Counts')
#    ax.legend()

plt.tight_layout()
plt.show();
```

```python
# Plot business hours vs. non-business hours and weekdays vs. weekends

fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Plot business hours vs. non-business hours
business_hours_agg = data.groupby('is_business_hour')['login_counts'].sum()
business_hours_agg.plot(ax=axs[0], color=['red', 'blue'], kind='bar', rot=0).set(xlabel=None)

axs[0].set_title('Total Logins for Business Hours vs. Non-Business Hours')
axs[0].set_xticklabels(['Non-Business Hours', 'Business Hours'])
axs[0].set_ylabel('Total logins')


# Plot business weekdays vs. weekends
weekend_agg = data.groupby('is_weekday')['login_counts'].sum()
weekend_agg.plot(ax=axs[1], color=['red', 'blue'], kind='bar', rot=0).set(xlabel=None)

axs[1].set_title('Total Logins for Weekend vs Weekday')
axs[1].set_xticklabels(['Weekend', 'Weekday']);

plt.tight_layout()
plt.show()
```

```python
# Plot boxplots for business hours vs. non-business hours and weekdays vs. weekends

# Plot mean for data in 15 min intervals
fig, axs = plt.subplots(1, 2, figsize=(15, 10))
sns.boxplot(x='is_business_hour', y='login_counts', data=data, ax=axs[0]).set(xlabel=None)
sns.boxplot(x='is_weekday', y='login_counts', data=data, ax=axs[1]).set(xlabel=None)

axs[0].set_xticklabels(['Non-Business Hours', 'Business Hours'])
axs[1].set_xticklabels(['Weekend', 'Weekday'])
axs[0].set_ylabel('Mean logins')
;
```

```python
daily_agg = data.groupby('day')['login_counts'].sum()

# Plot mean for data in hourly intervals
fig, axs = plt.subplots(1, 2, figsize=(15, 10))
sns.boxplot(x='is_business_hour', y='login_counts', data=data, ax=axs[0]).set(xlabel=None)
sns.boxplot(x='is_weekday', y='login_counts', data=data, ax=axs[1]).set(xlabel=None)

axs[0].set_xticklabels(['Non-Business Hours', 'Business Hours'])
axs[1].set_xticklabels(['Weekend', 'Weekday'])
axs[0].set_ylabel('Mean logins')
;
```

## Summary of Daily Cycles of Demand

There are several patterns that emerge when looking at the usage patterns for Ultimate Technologies Inc. is a transportation network company. A summary of the key findings is below:

#### 1. Demand increases during non-business hours and on weekends. 
From the bar charts and box plots, we can see that both the mean and total number of logins are higher on the weekends, and during non-business hours. This pattern is also evident when looking at the hourly patterns of daily usage.

#### 2. Demand increases during the lunch hour as well as during non-business hours.
There is a clear spike in demand during the lunch hour (around 11am-1pm) as well as during non-work hours. 

#### 3. Demand increases across months.
There is an increase in demand from January through April, with an apparent peak in March. This pattern is most apparent when looking at mean values (rather than total logins). This is an interesting pattern and worth exploring more -- perhaps demand increases during warmer months, although this depends on where Ultimate Tech transportation company is located.
